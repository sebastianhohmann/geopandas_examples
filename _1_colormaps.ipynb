# Colormaps in matplotlib

Often, when visualizing continuous variables, we want to show things using a color spectrum. 

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
plt.style.use('seaborn')
mpl.rcParams['font.family'] = 'serif'
```

`matplotlib` has wide range of built-in colormaps:
https://matplotlib.org/3.2.1/gallery/color/colormap_reference.html. We can get the explicit RGB-values of the different colors by first extracting the colormap and then  indexing the colormap at a point in $[0,1]$. For instance

```python
colors = mpl.cm.get_cmap('YlOrRd')
colors(0.3)
```

This allows us to plot objects with different colors at different discrete points of the colormap

```python
for i in np.linspace(0,1,10):
    plt.plot((0.1+i)*np.arange(10), color=colors(i))
```

We can also use only a subset of the range of the full colormap

```python
for i in np.linspace(0.3,0.6,10):
    plt.plot((0.1+i)*np.arange(10), color=colors(i))
```

We can also create our own colormap from the built-in colormap by focusing on a subset of the range

```python
rbg_newcmap = [colors(i) for i in np.linspace(0.3, 0.6, 10)]
newcmap = mpl.colors.ListedColormap(rbg_newcmap)
colors = mpl.cm.get_cmap(newcmap)
```

```python
for i in np.linspace(0,1,10):
    plt.plot((0.1+i)*np.arange(10), color=colors(i))
```

For more on how to make your own colormaps, see here: https://matplotlib.org/3.2.0/tutorials/colors/colormap-manipulation.html

The `seaborn` library has a very nice default color sequence.

```python
pal = sns.color_palette()
sns.palplot(pal)
```

Once againn, we can access the RGB-colors

```python
pal[0]
```

If we specify more than 6 colors, the colors begin to recyle

```python
pal10 = sns.color_palette(n_colors=10)
sns.palplot(pal10)
```

For lots of fancy examples and palettes with more than 6 colors, see here: https://seaborn.pydata.org/tutorial/color_palettes.html
